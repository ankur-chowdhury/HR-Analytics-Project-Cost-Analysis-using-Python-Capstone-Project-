```python
#--- TASK 1 ---
#There are three different tables as given above. Please make three dataframe in python and save them as three .csv files. From Task 2 to Task 10, use the saved .csv files only. 

import pandas as pd  # Import the pandas library to handle dataframes

# --- 1. Creating the Project DataFrame ---
# Data taken from the Project Table 
project_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005', 'A002', 'A005', 'A003', 'A001', 'A003', 'A001', 'A004', 'A004', 'A005'],
    'Project': ['Project 1', 'Project 2', 'Project 3', 'Project 4', 'Project 5', 'Project 6', 'Project 7', 'Project 8', 'Project 9', 'Project 10', 'Project 11', 'Project 12', 'Project 13', 'Project 14'],
    'Cost': [1002000, 2000000, 4500000, 5500000, None, 680000, 400000, 350000, None, 300000, 2000000, 1000000, 3000000, 200000],
    'Status': ['Finished', 'Ongoing', 'Finished', 'Ongoing', 'Finished', 'Failed', 'Finished', 'Failed', 'Ongoing', 'Finished', 'Failed', 'Ongoing', 'Finished', 'Finished']
}
df_project = pd.DataFrame(project_data) # Converting the dictionary into a pandas DataFrame
df_project.to_csv('Project.csv', index=False) # Save as Project.csv 

# --- 2. Creating the Employee DataFrame ---
# Data taken from the Employee Table
employee_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Name': ['John Alter', 'Alice Luxumberg', 'Tom Sabestine', 'Nina Adgra', 'Amy Johny'],
    'Gender': ['M', 'F', 'M', 'F', 'F'],
    'City': ['Paris', 'London', 'Berlin', 'Newyork', 'Madrid'],
    'Age': [25, 27, 29, 31, 30]
}
df_employee = pd.DataFrame(employee_data) # Converting dictionary to DataFrame
df_employee.to_csv('Employee.csv', index=False) # Save as Employee.csv 

# --- 3. Creating the Seniority Level DataFrame ---
# Data taken from the Seniority Level Table

seniority_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Designation Level': [2, 2, 3, 2, 3]
}
df_seniority = pd.DataFrame(seniority_data) # Convert dictionary to DataFrame
df_seniority.to_csv('Seniority.csv', index=False) # Save as Seniority.csv 

print("Task 1 complete: All three CSV files have been created.") # Confirmation message
```

```python
#--- TASK 2 ---
#The cost column in the dataframe “Project” has some missing values. Your task is to compute these missing values. Replace the missing values by running average. You should use the “For” loop for this task.

import pandas as pd  # Import pandas to work with the dataframe

# Loading the Project CSV file created in Task 1
df_project = pd.read_csv('Project.csv') # Reading the saved file 

# Initializing variables to keep track of the sum and the count of valid numbers

total_sum = 0 # To store the sum of costs encountered so far
count = 0     # To store the number of projects with valid costs encountered so far

# Using a For loop to iterate through the 'Cost' column 
for i in range(len(df_project)):     # Looping through every row index in the dataframe
    cost = df_project.loc[i, 'Cost'] # Get the cost value at the current row
    
    # Checking if the cost is a valid number (not NaN/missing)
    if pd.notnull(cost): 
        total_sum += cost # Add current cost to the total sum
        count += 1        # Increase the count of valid costs
    else:
        # If the cost is missing, we need to calculate the running average 
        if count > 0:                               # Prevent division by zero
            running_avg = total_sum / count         # Calculate average: Total Sum / Number of items
            df_project.loc[i, 'Cost'] = running_avg # Replacing the missing value 
            
            # Since we just filled a gap, we update the sum and count for the next iteration
            total_sum += running_avg 
            count += 1

# Displaying the updated dataframe to see the filled values
print("Task 2 complete: Missing costs filled with running average.")
print(df_project)
```

```python
#--- TASK 3 ---
#Split the name column in the Employee dataframe into two new columns “First Name”, and “Last Name” and remove the older “name” column. 


import pandas as pd # Import pandas to handle the data

# Loading the Employee CSV file created in Task 1
df_employee = pd.read_csv('Employee.csv') # Reading the file

# Split the 'Name' column into two parts based on the space between names
# 'expand=True' puts the results into two separate new columns
df_employee[['First Name', 'Last Name']] = df_employee['Name'].str.split(' ', expand=True)
# This deletes the old column. Here inplace=True means "applying this change directly to the table I'm working on."

# Removing the original 'Name' column as it is no longer needed
df_employee.drop(columns=['Name'], inplace=True)

# Display the result to verify the split
print("Task 3 complete: Name column split into First Name and Last Name.")
print(df_employee.head()) # Shows the first few rows of the updated table
```

```python
#--- TASK 4 ---
#Join all three dataframes in one single dataframe. Name it “Final”.


import pandas as pd  # Import pandas for data manipulation

# 1. Load the data from Task 2 (Project) and Task 3 (Employee)
# We need to load the Seniority file we created in Task 1.
df_seniority = pd.read_csv('Seniority.csv') # Loading the seniority level data

# 2. Join the Project and Employee dataframes first
# We use 'merge' on the 'ID' column because it exists in both tables.
# 'how=inner' ensures we only keep rows where the ID matches in both.
merged_temp = pd.merge(df_project, df_employee, on='ID', how='inner')

# 3. Join the result with the Seniority dataframe to include Designation Level
# We store this in a new variable called 'Final' as requested.
Final = pd.merge(merged_temp, df_seniority, on='ID', how='inner')

# 4. Displaying the first few rows to confirm the join worked
print("Task 4 complete: All dataframes joined into 'Final'.")
print(Final.head()) # Showing the combined table
```

```python
#--- TASK 5 ---
#Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project cost only to employees who have finished the projects. 

import numpy as np # Import numpy for conditional calculations

# We need to create a new column 'Bonus'
# Logic used: If 'Status' is 'Finished', Bonus = Cost * 0.05. Otherwise, Bonus = 0.
Final['Bonus'] = np.where(Final['Status'] == 'Finished', Final['Cost'] * 0.05, 0)

# Displaying the table to see the new Bonus column
print("Task 5 complete: 5% bonus calculated for Finished projects.")
print(Final[['ID', 'Project', 'Status', 'Cost', 'Bonus']].head(10))
```

```python
#--- TASK 6 ---
#Demote the designation level by 1, whose projects have status “fail”. Delete the employees record whose designation level is above 4. 

#--- Step 1: Demote based on 'Failed' status ---
# Using .loc to find rows where Status is 'Failed' and add 1 to the 'Designation Level'
# Every line must have comments as per requirements
Final.loc[Final['Status'] == 'Failed', 'Designation Level'] += 1 

# --- Step 2: Delete records with Designation Level > 4 ---
# We are redefining 'Final' to only include rows where the Level is 4 or less
# This will effectively deletes anyone who crossed the eligibility threshold 

Final = Final[Final['Designation Level'] <= 4] 

# Displaying the results to check for changes and output
print("Task 6 complete: Demotions applied and ineligible records removed.") 
print(Final[['ID', 'Project', 'Status', 'Designation Level']])
```

```python
#--- TASK 7 ---
#Add “Mr.” and “Mrs.” to the first name column and drop the gender column.


import pandas as pd
import numpy as np

# Run the logic if 'Gender' column exists

if 'Gender' in Final.columns:
    # Step 1: We need to be certain that the 'First Name' is treated as text
    Final['First Name'] = Final['First Name'].astype(str)

    # Step 2: Adding Mr./Mrs. based on Gender column
    # If Gender is 'M' it adds 'Mr. ', otherwise 'Mrs. '
    Final['First Name'] = Final.apply(lambda row: 'Mr. ' + row['First Name'] if row['Gender'] == 'M' else 'Mrs. ' + row['First Name'], axis=1)

    # Step 3: Dropping the Gender column
    Final.drop(columns=['Gender'], inplace=True)
    print("Task 7 complete: Titles added and Gender column dropped.")
else:
   
    print("Task 7 was already completed. 'Gender' column is no longer in the table.")

# Displaying the results
print(Final[['ID', 'First Name', 'Last Name']].head()) #Dispaying only required columns
```

```python
#--- TASK 8 ---
#Promote designation level by 1 for the employees whose age is more than 29 years using IF condition.

# We loop through the index of the 'Final' table one row at a time
for i in Final.index: 
    
    # 1. Checking the age and current designation for the current row
    
    age = Final.loc[i, 'Age']
    current_level = Final.loc[i, 'Designation Level']
    
    # 2. Applying the 'IF' condition to check: Is the employee older than 29?
    if age > 29:
        
        # Promote by 1 (Subtracting 1 makes the rank higher on a 1-4 scale)
        new_level = current_level - 1
        
        # 3. Ensuring the level doesn't go below 1 (Because 1 is the highest rank)
        if new_level < 1: 
            new_level = 1
            
        # 4. Updating the table with the new promoted level
        Final.loc[i, 'Designation Level'] = new_level

# Displaying the updated details for verification
print("Task 8 complete: Promotions applied to employees over 29 using IF condition.")
print(Final[['ID', 'First Name', 'Age', 'Designation Level']].head())
```

```python
#--- TASK 9 ---
#Add the cost of all projects for each Employee and save it in new dataframe “TotalProjCost” with three columns ID, First Name, and Total cost.

# 1. First we need to group the 'Final' table by 'ID' and 'First Name'
# 2. Then we need to select the 'Cost' column and use .sum() to add all costs for those specific people
# 3. .reset_index() turns the result back into a clean table

TotalProjCost = Final.groupby(['ID', 'First Name'])['Cost'].sum().reset_index()

# 4. Renaming the 'Cost' column to 'Total cost'
TotalProjCost.rename(columns={'Cost': 'Total cost'}, inplace=True)

# Displaying the new summary table
print("Task 9 complete: Total project costs calculated per employee.")
print(TotalProjCost)
```

```python
#--- TASK 10 ---
#Print all the employee details whose city name contains the letter “o” in it.

# 1. We need to start a loop to look at every row in our 'Final' table one by one

for i in Final.index:
    # 2. Getting the city name for the current row "city_name"
    
    city_name = Final.loc[i, 'City'].lower()
    # We use .lower() to convert the city name in lower case, to make sure we find 'o' even if it's written as 'O'
    
    # 3. Checking if 'o' is inside the city name
    if 'o' in city_name:
        
        # 4. If 'o' is found, printing all the details for that specific employee row
        print("Employee Found in City with 'o' letter:")
        print(Final.loc[i])
        print("-" * 30)  # This is just to print a line to separate each person details
```
